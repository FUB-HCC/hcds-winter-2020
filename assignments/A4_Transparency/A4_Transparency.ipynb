CONTEXT ->

1. Edit quality - review of potentially damaging contributions
* `reverted` – predicts whether an edit will eventually be reverted
* `damaging` – predicts whether or not an edit causes damage
* `goodfaith` – predicts whether an edit was saved in good-faith

2. Article quality
* Curation support: `draftquality` – predicts if the article will need to be speedy deleted (spam, vandalism, attack, or OK)
* Assessment scale support: `articlequality` – predicts the (Wikipedia 1.0-like) assessment class of an article or draft

3. Topic routing
ORES topic models (word embeddings)

* Curation support: `drafttopic` – predicts the topic of an a new article draft
* Topic interest mapping: `articletopic` – predicts the topic of an article (more details)

## How familiar are you with the following topics?

```python
ldata = df[df.columns[df.columns.str.contains(pat = 'How')]] #ldata => Likert Scale Data
```

```python
ldata.head(5)
```

Lets again remove all rows with all NaN.

```python
ldata = ldata.dropna(how='all')
```

```python
ldata.tail(5)
```

## Basic diverging stacked bar graphs for visualising likert scale data
* https://blog.orikami.nl/behind-the-screens-likert-scale-visualization-368557ad72d1
* https://blog.orikami.nl/behind-the-screens-likert-scale-visualization-368557ad72d1

```python
ldata = ldata.replace("Extremely familiar", 1) #1
ldata = ldata.replace("Very familiar", 2) #2
ldata = ldata.replace("Moderately familiar", 3) #2
ldata = ldata.replace("Slightly familiar", 4) #4
ldata = ldata.replace("Not at all familiar", 5) #5
ldata = ldata.fillna(0)
```

```python
ldata
```

```python
data_pivot = ldata.apply(lambda x: x.value_counts())
```

```python
data_pivot
```

```python
data_pivot = data_pivot.fillna(0)
```

```python
data_pivot
```

```python
data_pivot.plot(kind="barh", stacked=True)
```

```python
data_pivot = data_pivot.transpose()
```

```python
data_pivot
```

```python
data_pivot.plot(kind="barh", stacked=True)
```

```python
# final_data = data_pivot.drop(columns=[0])
final_data = data_pivot
```

```python
final_data
```

```python
final_data.plot(kind="barh", stacked=True)
```
